```python
#checking for null values
len(df) - len(df.dropna())
```

```python
nan_dataf = df[df.isna().any(axis=1)]
display(nan_dataf.head())
```

# Finding the impact of marital status of a user impact in spending on wine

```python
dm=df.groupby(['Marital_Status']).mean()
dm
```

```python
dm['Mindex']=['Absurd','Alone','Divorced','Married','Single','Together','Widow','YOLO']
```

```python
plt.figure(figsize=(9,6))
plt.title('Impact of Marital Status on Wine Expenses')
plt.bar(dm.Mindex,dm.MntWines)
plt.xlabel('Marital_Status')
plt.ylabel('Money spent on wine')
plt.grid()
```

```python
#thus by analyzing the graph we can say that being widow and absurd makes a huge impact on wine expenses
```

# If a user has done more higher education does he spend more?

```python
de=df.groupby(['Education']).aggregate({'TotalMntSpent':[np.median, np.mean, max]})
de
```

### Thus by looking at the numbers we can see that customers who have done   PhD and graduation are on an average spending more money

### But by looking at the max value we can see there is a special case where the highest expense is done by  customer with a Masters degree

```python
#the customers with highest expenses with Masters degree
df.groupby('TotalMntSpent').get_group(2525)
```

# Users who have kids at home are likely to spend more money in gold?

```python
df['Kidhome'].unique()
```

```python
dg=df.groupby(['Kidhome']).mean()
```

```python
dg['MntGoldProds']
```

```python
plt.figure(figsize=(6,9))
plt.title('Impact on gold expenses for having kids in house')
mylabels = df['Kidhome'].unique()
plt.pie(dg.MntGoldProds,labels= mylabels)
plt.xlabel('no of kids')
```

### thus we can conclude the having kids at home decreases gold purchases
